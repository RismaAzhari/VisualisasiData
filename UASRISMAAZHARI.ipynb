```python
import pandas as pd

#reading the database
data = pd.read_csv("/content/kebutuhan_per_bulan_2.csv")

#printing the top 10 rows
display(data.head(10))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

x = data['tahun']
y = data['Propinsi']

plt.scatter(x,y)

plt.plot(x,y, color ='g', linestyle ='dashed')

plt.xticks(rotation = 25)
plt.xlabel('Tahun')
plt.ylabel('Propinsi')
plt.title('kebutuhan per bulan', fontsize = 12)
plt.grid()

plt.show()

```

```python
import pandas as pd

# Membaca dataset profil pelanggan dari file CSV
data = pd.read_csv('SupplyKantor.csv')

# Mencetak 10 baris pertama dari dataset
print(data.head(10))
```
